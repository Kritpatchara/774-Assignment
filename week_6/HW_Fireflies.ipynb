```python
import numpy as np
import matplotlib.pyplot as plt

# Setting constant
omega_1 = np.pi
omega_2 = 6*np.pi
A = 2*np.pi
dt = 0.0001 # (time step)

# Initial condition
theta_1 = [0]
theta_2 = [np.pi]
t=[0]

for i in range(0, 50000):
    theta1 = (((omega_1 + (A*np.sin(theta_2[i]-theta_1[i])))*dt)+theta_1[i]) % (2*np.pi)
    theta_1.append(theta1)
    
    theta2 = (((omega_2 + (A*np.sin(theta_1[i]-theta_2[i])))*dt)+theta_2[i]) % (2*np.pi)
    theta_2.append(theta2)
    
    t1=t[i]+dt
    t.append(t1)

plt.plot(t,theta_1)
plt.plot(t,theta_2)
plt.show()

# "phase locking" = Different between phase is constant
```

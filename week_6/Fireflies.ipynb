```python
import numpy as np
import matplotlib.pyplot as plt

## No interaction (Asin(Theta-theta))

# omega = 1Hz = 2pi rad/s
# initial condition: theta = 0, Theta = pi
# Give A = 2pi
# Set experiment lide Ermentrout: Omega = 1Hz (2pi rad/s), 1.5Hz (3pi rad/s), 3Hz (6pi rad/s)

# Setting constant
omega=2*np.pi
Omega=2*np.pi
A=2*np.pi
dt=0.001 # (time step)

# Initial condition
theta=[0]
Theta=[np.pi]
t=[0]

for i in range(0,10001): # Simulate 10 000 times (10 sec)
    theta1=((omega*dt)+theta[i]) % (2*np.pi) # mod because phase 2pi = 0 =
    theta.append(theta1)
    
    Theta1=((Omega*dt)+Theta[i]) % (2*np.pi)
    Theta.append(Theta1)
    
    t1=t[i]+dt
    t.append(t1)

plt.plot(t,theta)
plt.plot(t,Theta)
plt.show()

# blue = fireflies, orange = light
```

```python
## Have an interaction (Asin(Theta-theta))

omega=2*np.pi
Omega=3*np.pi
A=2*np.pi
dt=0.001

theta=[0]
Theta=[np.pi]
t=[0]

for i in range(0,10001):
    theta1=(((omega+(A*np.sin(Theta[i]-theta[i])))*dt)+theta[i]) % (2*np.pi)
    theta.append(theta1)
    
    Theta1=((Omega*dt)+Theta[i]) % (2*np.pi)
    Theta.append(Theta1)
    
    t1=t[i]+dt
    t.append(t1)

plt.plot(t,theta)
plt.plot(t,Theta)
plt.show()

# "phase locking" = Different between phase is constant
```

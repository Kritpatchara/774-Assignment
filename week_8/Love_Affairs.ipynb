```python
import numpy as np
import matplotlib.pyplot as plt

# Setting constant
a = 1
b = 1
dt = 0.01 # (time step)

# Initial condition
R=[10]
J=[10]
t=[0]

for i in range(0,1000): # Simulate 10 000 times (10 sec)
    Ri = a*J[i]*dt+R[i]
    R.append(Ri)
    
    Ji = -b*R[i]*dt+J[i]
    J.append(Ji)
    
    t1=t[i]+dt
    t.append(t1)

y = 0
plt.plot(t,R)
plt.plot(t,J)
plt.plot(y)
plt.show()

# blue = Romeo, orange = Juliet
```

```python
import numpy as np
import matplotlib.pyplot as plt

R = [10]
J = [5]
a = -10
b = 5
dt = 0.01

for i in range(0,100):
    
    R1 = ((a * R[i] + b * J[i]) * dt) + R[i]
    J1 = ((b * R[i] + a * J[i]) * dt) + J[i]
    
    R.append(R1)
    J.append(J1)
    
plt.plot(R)
plt.plot(J)
plt.show()
```

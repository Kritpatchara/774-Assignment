```python
import numpy as np
import matplotlib.pyplot as plt

# Setting parameter
A = -0.5
dt = 0.01  # (time step)

# Initial condition
N = [1000]
t = [0]

for i in range(0, 2000):
    Ni = (A*N[i]*dt+N[i])
    N.append(Ni)

    t1 = t[i]+dt
    t.append(t1)

plt.plot(t, N)
plt.show()
print(A*dt+1)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Setting parameter


# Initial condition
x = [0.75] # of carrying capacity
mu = 4*(1+np.finfo(float).eps)

for i in range(0, 50):
    x1 = (x[i]*mu*(1-x[i]))
    x.append(x1)


plt.plot(x)
plt.show()
```
